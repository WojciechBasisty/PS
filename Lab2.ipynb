```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.integrate import quad
```

```python
T=2
a=-1
b=1
w=2*np.pi/T
p=1+18/100
#bn=0 bo funkcja parzysta
```

```python
def f(t):
  return p*t**2
def f1(t):
  return f(t)*np.cos(w*t)
def f2(t):
  return f(t)*np.cos(2*w*t)

a0=(1/T) *quad(f,a,b)[0]
a1=(1/T) *quad(f1,a,b)[0]
a2=(1/T) *quad(f2,a,b)[0]

print ("p =",p ,"a0 =", a0 , "a1 =", a1 , "a2 =", a2 )
```

```python
def sfinal(t):
  return (a0/2)+a1*np.cos(w*t)+a2*np.cos(2*w*t)
```

```python
t = np.linspace(a,b, 100)
tf = np.linspace(-3,3, 100)
fig, ax = plt.subplots()

ax.plot(t,f(t),tf,sfinal(tf))

ax.grid(True)

plt.show()
```

```python
k=10

yb=p*b**2
cal=[]
for x in np.arange(-k,k,0.1):
  i=0
  y=p*x**2
  if x<a:
    while y>yb:
      i+=T
      y=p*(x+i)**2
  elif x>=b:
    while y>yb:
      i-=T
      y=p*(x+i)**2
  cal.append(y)


t = np.arange(-k,k, 0.1)

fig, ax = plt.subplots()

ax.plot(t,cal)

ax.grid(True)

plt.show()
```
